```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset from a text file
data_path = r'C:\Users\ebenr\OneDrive\Documents\dataset automobile\automobile.txt'
column_names = ['symboling', 'normalized-losses', 'make', 'fuel-type', 'aspiration', 'num-of-doors',
                'body-style', 'drive-wheels', 'engine-location', 'wheel-base', 'length', 'width', 'height',
                'curb-weight', 'engine-type', 'num-of-cylinders', 'engine-size', 'fuel-system', 'bore',
                'stroke', 'compression-ratio', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
df = pd.read_csv(data_path, names=column_names)

# Remove duplicate rows
df = df.drop_duplicates()

df = df[df['price'] != '?'].copy()

# Discard entries with zero in numeric columns
numeric_columns = ['normalized-losses', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-size',
                   'bore', 'stroke', 'compression-ratio', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')
df[numeric_columns] = df[numeric_columns].replace('?', pd.NA)
df = df.dropna(subset=numeric_columns)

for col in numeric_columns:
    if col != 'normalized-losses':  # Skip the column that caused the error
        df[col] = df[col].astype(float)
        
# Convert certain columns to appropriate data types
df[numeric_columns] = df[numeric_columns].astype(float)

# Convert 'num-of-doors' to numeric values
door_mapping = {'two': 2, 'four': 4}
df['num-of-doors'] = df['num-of-doors'].map(door_mapping)

# Convert 'num-of-cylinders' to numeric values
cylinder_mapping = {'two': 2, 'three': 3, 'four': 4, 'five': 5, 'six': 6, 'eight': 8, 'twelve': 12}
df['num-of-cylinders'] = df['num-of-cylinders'].map(cylinder_mapping)

# Print the first few rows of the cleaned DataFrame
print(df.head())

# Analyze the data and look for trends
# For example:
# Average price by make
average_price_by_make = df.groupby('make')['price'].mean()
print("\nAverage Price by Make:")
print(average_price_by_make)

# Create a bar plot for the frequency of body styles
body_style_counts = df['body-style'].value_counts()

# Visualize the frequency of body styles
plt.figure(figsize=(10, 6))
sns.barplot(x=body_style_counts.index, y=body_style_counts.values)
plt.xticks(rotation=45)
plt.xlabel('Body Styles')
plt.ylabel('Number of Vehicles')
plt.title('Frequency of Vehicles in Each Body Style')
plt.show()

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = {
    'symboling': [2, 2, 1, 1, 2],
    'normalized-losses': [164.0, 164.0, 158.0, 158.0, 192.0],
    'make': ['audi', 'audi', 'audi', 'audi', 'bmw'],
    'fuel-type': ['gas', 'gas', 'gas', 'gas', 'gas'],
    'aspiration': ['std', 'std', 'std', 'turbo', 'std'],
    'num-of-doors': [4.0, 4.0, 4.0, 4.0, 2.0],
    'body-style': ['sedan', 'sedan', 'sedan', 'sedan', 'sedan'],
    'drive-wheels': ['fwd', '4wd', 'fwd', 'fwd', 'rwd'],
    'engine-location': ['front', 'front', 'front', 'front', 'front'],
    'wheel-base': [99.8, 99.4, 105.8, 105.8, 101.2],
    'engine-size': [109.0, 136.0, 136.0, 131.0, 108.0],
    'price': [13950.0, 17450.0, 17710.0, 23875.0, 16430.0]
}

df = pd.DataFrame(data)

# Create a bar plot for Average Price by Make
average_price_by_make = df.groupby('make')['price'].mean()
plt.figure(figsize=(10, 6))
sns.barplot(x=average_price_by_make.index, y=average_price_by_make.values)
plt.xticks(rotation=45)
plt.xlabel('Make')
plt.ylabel('Average Price')
plt.title('Average Price by Make')
plt.tight_layout()
plt.show()

# Create a bar plot for Body Style Counts
body_style_counts = df['body-style'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=body_style_counts.index, y=body_style_counts.values)
plt.xticks(rotation=45)
plt.xlabel('Body Styles')
plt.ylabel('Number of Vehicles')
plt.title('Frequency of Vehicles in Each Body Style')
plt.tight_layout()
plt.show()

```
